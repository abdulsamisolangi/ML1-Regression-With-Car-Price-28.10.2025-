```python
class Student:      #step 2
    def __init__(self,name=None,section=None,percentage=None,age=None):  #constructor
        self.name = input("Enter Your name:")      #attribute
        self.section = input("Enter Your section:")    #attribute
        self.percentage = float(input("Percentage:"))
        self.age = input("Enter Your age:")
        
    def Pass_fail(self):
        return "Pass" if self.percentage>40 else "Fail"
        
    def student_details(self):     #method
        st=[self.name,self.section,self.percentage,self.age]
        print(st)
```

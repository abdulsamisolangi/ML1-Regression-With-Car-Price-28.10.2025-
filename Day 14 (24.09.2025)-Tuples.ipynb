## SETS

```python
#characteristics of set
#1: Unique Values / items.
#2: Unordered - no indexing.
#3: mutable - add / remove elements.
```

```python
my_set={1,2,3,4,5}
print(type(my_set))
```

```python
my_set2=set([6,7,8,9,10])
print(my_set2)
```

```python
my_set2=set([6,7,8,9,10])
print(my_set2)
my_set3=set([])
print(type(my_set3))
```

```python
print(my_set)
my_set2
```

```python
my_set.add(100)
print(my_set)
```

```python
my_set.remove(100)
print(my_set)
```

```python
#discard:
my_set.discard(4)
print(my_set)
```

```python
#Set

#1. Union
#2. Intersection
```

```python
#Union:

set1={1,2,3,4,5,6}
set2={6,7,8,9,10}

print(set1.union(set2))
```

```python
set1 | set2 #Union alternative:
```

```python
set1.intersection(set2)
```

```python
set1 & set2 #intersection alternatvie
```

```python
#to check difference
set1.difference(set2)
```

```python
set2.difference(set1)
```

```python
set1 - set2  #differences alternative
```

```python
#Symmetric differences:
set1.symmetric_difference(set2)
```

```python
set1 ^ set2 #alternative
```

```python
for x in set1:
    print(x)
```

```python
#Squares:
squ = {x**4 for x in range(6)}
print(squ)
```

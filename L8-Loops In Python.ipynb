```python
#if True Loop:
#while True Loop:

#Example:
guess_number="0"
secret_number="7"
if True:
    guess_number=input("guess the number 1 to 7:")
    if guess_number == secret_number:
        print("Yes", guess_number, "is correct")
    else:
        print(guess_number, "is incorrect")
```

```python
#Example:
guess_number="0"
secret_number="7"
while True:
    guess_number=input("guess the number 1 to 7:")
    if guess_number == secret_number:
        print("Yes", guess_number, "is correct")
        break   #(break is always used in while True otherwise the loop continues until condition is True).
    else:
        print(guess_number, "is incorrect")
```

```python
#for loop:
. A for loop is used to iterate over of sequence (such as list, tuple or string).
. Use for loop when you know exactly, how many times you  want to execute the loop.
        
#Example:
colors=("Blue", "Black", "Green", "Red")
for i in colors:
    print(i)
```

```python
for i in range(1,10):
    print(i)
```

```python
for i in range(4):
    x=int(input("Enter a number between 1 and 100:"))
    if 1<=x<=100:
        print("valid no", x)
        break
    else:
        print("valid number, try again")
```

```python
#while loop:
. A while loop is use to repeatedly execute a block of code as long as a certain condition is true.
. Use while loop when you do not know how many times you execute the loop beforehand.
```

```python
Example:
i=1
while i<5:
    print(i)
    i +=1 #(+ is increment)
```

```python
x=0
while not(1<=x<=100):
    x=int(input("Enter a number between 1 and 100:"))
    print("valid no", x)
```

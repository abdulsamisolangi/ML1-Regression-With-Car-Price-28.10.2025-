# Object Oriented Programming in Python:

```python
class Student2:
    def __init__(self):   #Constructor
        self.name=input("Enter Your name:")    #attributes
        self.age=input("Enter Your age:")      #attributes
        self.grade=input("Enter Your grade:")  #attributes
    
    def Student2_details(self):
        print(f"name: {self.name}, and age is {self.age}, and class is {self.grade}")
        
    def per(self):
        phy=int(input("Enter Physics Marks:"))
        chem=int(input("Enter Chemistry Marks:"))
        pst=int(input("Enter Pak.Studies Marks:"))
        maths=int(input("Enter Mathmatics Marks:"))
        isl=int(input("Enter Islamiat Marks:"))
        obtain=phy+chem+pst+maths+isl
        per=(obtain/500)*100

        if per>=80:
            grade= "A+"
        elif per>=70:
            grade= "A"
        elif per>=60:
            grade= "B"
        elif per>=50:
            grade= "C"
        else:
            grade= "Fail"
        return f"Your Grade is {grade} and Your percentage is {per}"
```

```python
st1=Student2()
```

```python
st1.per()
```

```python
#Abstraction:

#Hiding Unnecessary details from users through class, method.

class Student1:
    def __init__(self):   #Constructor
        self.name=input("Enter Your name:")    #attributes
        self.age=input("Enter Your age:")      #attributes
        self.grade=input("Enter Your grade:")  #attributes
        self.obtain=int(input("Enter your obtained Mmarks:"))
        self.percentage=(self.obtain/500)*100
    def Student1_details(self):
        print(f"name: {self.name}, and age is {self.age}, and class is {self.grade}")
```

```python
st1=Student1()
```

```python
st1.Student1_details()
```

```python
#Encapsulation:

#Restrict access to certain attribute or methods to protect data and enforce controlled access.

class Student2:
    def __init__(self):   #Constructor
        self.name=input("Enter Your name:")    #attributes
        self.age=input("Enter Your age:")      #attributes
        self.grade=input("Enter Your grade:")  #attributes
        self.obtain=int(input("Enter your obtained Marks:"))
        self.__percentage=(self.obtain/500)*100    #Double underscore (__) is used to limit the access of value.
    def student_details2(self):
        print(f"name: {self.name}, and age is {self.age}, and class is {self.grade}")
    def get_per(self):
        return self.__percentage+5       
```

```python
st5=Student2()
```

```python
st5.get_per()
```

```python
st5.student_details2()
```

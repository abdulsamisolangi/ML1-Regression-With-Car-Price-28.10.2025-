## Lets clean and remove spaces from name

```python
df['Manager']=df['Manager'].str.strip().str.replace(r'\s+',' ', regex=True)

#.str.strip()
#removes any leading and trailing space from each string

#.str.replace(r'\s+',' ', regex=True)

#uses a regular exp. (r'\s+') to replace one or more 
#white spaces characters with a single space so if the name has multiple spaces b/w words,
#they get replaced to one!


#Tom    jackson ---> Tom jackson

#regex 

#---- Data cleaning (removing unwanted space, symbols etc)
#---- Searching Text (finding specific word)
#---- Validations
#---- Text extraction
```

## City Wise Sale:

```python
df['City']
```

```python
#Step 1:
df.groupby('City')['Sales'].sum().plot(
    kind='pie',
    autopct='%1.2f%%',
    colors=['red','blue','green','orange','gray'],
    wedgeprops={'linewidth':2,'edgecolor':'white'}
)

plt.title("City Wise Sales")
```

```python
#Step 2:
df.groupby('City')['Sales'].sum().plot(kind='pie',autopct='%1.2f%%',colors=['red','blue','green','orange','pink'])
plt.title("City Wise Sales") #plt.title=for title
```

```python
#Step 3 with some additional features:
df.groupby('City')['Sales'].sum().plot(
    kind='pie',
    autopct='%1.2f%%',
    colors=['red','blue','green','orange','gray'],
    wedgeprops={'linewidth':2,'edgecolor':'white'},
    pctdistance=1.25, #pct = percentage
    labeldistance=.6,
    startangle=45,
    explode=(0,0.1,0,0,0)
)

plt.title("City Wise Sales")
```

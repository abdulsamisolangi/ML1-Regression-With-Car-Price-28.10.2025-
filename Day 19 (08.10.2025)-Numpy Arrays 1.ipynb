```python
arr7 = np.array([10,2,3,4,5,78,3])

print(arr7.sum())

sine_values = np.sin(np.radians(arr7))
cosine_values = np.cos(np.radians(arr7))
square_root = np.sqrt(arr7).round(2)   #.round = if you want to restrict decimal
exponential = np.exp(arr7).round(2)

print(sine_values, "\nCosine Value",cosine_values, "\nSquare Root",square_root, "\nExpon",exponential)
```

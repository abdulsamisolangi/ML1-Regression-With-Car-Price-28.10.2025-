## Q1: Quantity of Products Regionwise:

```python
cs.columns
```

```python
cs.info()
```

```python
cs['Product'].value_counts()
```

```python
cs.groupby(['Region','Product'])['Qty'].sum()
```

```python
cs['Sales']= cs['Price'] * cs['Qty']
```

```python
cs
```

```python
cs.groupby('Region')[['Sales','Qty']].sum().reset_index() #psq=Product Sales Quantity
```

```python
!pip install matplotlib
```

```python
import matplotlib.pyplot as plt
```

```python
cs.groupby('Region')['Sales'].sum().plot(
    kind='pie',
    autopct='%1.2f%%',
    colors=['red','blue','green','orange'],
    wedgeprops={'linewidth':2,'edgecolor':'white'}
)

plt.title("Region Wise Sales")
```

## Q2: Year Wise Sales alongwith region:

```python
cs.groupby('Year')[['Region','Sales']].count().reset_index()
```

```python
cs.groupby('Year')['Sales'].sum().plot(kind='pie',autopct='%1.2f%%',colors=['Yellow','blue','Maroon'])
plt.title("Year Wise Sales") #plt.title=for title
```

## Q3: Top 2 Categories Wise Sale:

```python
cs['Category']
```

```python
cs.groupby('Category')[['Sales','Qty']].sum().reset_index()
```

```python
cs.groupby('Category')['Sales'].sum().plot(
    kind='pie',
    autopct='%1.2f%%',
    colors=['red','skyblue','pink','orange','gray'],
    wedgeprops={'linewidth':2,'edgecolor':'white'},
    pctdistance=1.25, #pct = percentage
    labeldistance=.6,
    startangle=45,
    explode=(0,0.1,0,0,0)
)

plt.title("Category Wise Sales")
```

## Q4: Total Investment on Products (Dataframe) Visualize:

```python
cs.groupby('Product')[['Price','Amount']].sum().reset_index()
```

```python
cs['Amount'].sum()
```

```python
cs.plot(kind='bar') 
plt.xlabel('Sales')
plt.ylabel('Amount')

plt.show()
```
